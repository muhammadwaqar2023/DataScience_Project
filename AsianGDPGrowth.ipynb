```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(color_codes=True)
```

```python
dataset = pd.read_csv('gdpdataset.csv')

dataset
```

```python
#Indicate duplicates
duplicate_rows_dataset = dataset[dataset.duplicated()]
print("Number of Duplicate Rows: ", duplicate_rows_dataset.shape)
```

```python
#Remove duplicates
dataset = dataset.drop_duplicates()
dataset.head(5)
```

```python
# Check the current columns
print("Columns before removing 'Indicator Code':", dataset.columns)
```

```python
# Remove the 'Indicator Code' column
dataset = dataset.drop(columns=['Indicator Code'])
```

```python
# Check the columns after removal
print("Columns after removing 'Indicator Code':", dataset.columns)
```

```python
# Save the updated dataset to a new CSV file if needed
dataset.to_csv('updated_gdpdataset.csv', index=False)
```

```python
regions = dataset[dataset['Country Name'].isin(['Sub-Saharan Africa','East Asia', 'South Asia', 'Europe & Central Asia', 'Middle East & North Africa','Latin America & Caribbean'])]

regions
```

```python
Asian_countries = dataset[dataset['Country Name'].isin(['Afghanistan', 'Armenia', 'Azerbaijan', 'Bahrain', 'Bangladesh', 'Bhutan', 'Brunei', 'Cambodia', 'China',
                  'Cyprus', 'Georgia', 'India', 'Indonesia', 'Iran', 'Iraq', 'Israel', 'Japan', 'Jordan', 'Kazakhstan',
                   'Kuwait', 'Kyrgyzstan', 'Laos', 'Lebanon', 'Malaysia', 'Maldives', 'Mongolia', 'Myanmar (Burma)', 'Nepal',
                   'North Korea', 'Oman', 'Pakistan', 'Palestine', 'Philippines', 'Qatar', 'Saudi Arabia', 'Singapore',
                   'South Korea', 'Sri Lanka', 'Syria', 'Taiwan', 'Tajikistan', 'Thailand', 'Timor-Leste', 'Turkey',
                   'Turkmenistan', 'United Arab Emirates', 'Uzbekistan', 'Vietnam', 'Yemen'])]

top_10_asian_countries_2022 = Asian_countries.sort_values(by='2022', ascending=False).head(10)

top_10_asian_countries_2022

```

```python
# Extracting relevant data for the pie chart
countries = top_10_asian_countries_2022['Country Name']
values = top_10_asian_countries_2022['2022']

# Creating a pie chart
plt.figure(figsize=(10, 6))
plt.pie(values, labels=countries, autopct='%1.1f%%', startangle=140)
plt.title('Top 10 Asian Countries in 2022')
plt.show()
```

```python
# Create a bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='2022', y='Country Name', data=top_10_asian_countries_2022, palette='viridis')

# Set plot labels and title
plt.xlabel('2022')
plt.ylabel('Country Name')
plt.title('Top 10 Asian Countries in 2022')

# Show the plot
plt.legend()
plt.show()
```

```python
# Create a combined plot with bars and a line
plt.figure(figsize=(10, 6))

# Bar plot
sns.barplot(x='Country Name', y='2022', data=top_10_asian_countries_2022, color='blue', alpha=0.7, label='2022 Data')

# Line plot
sns.lineplot(x='Country Name', y='2022', data=top_10_asian_countries_2022, marker='o', color='green', label='Line Plot')

# Set plot labels and title
plt.xlabel('Country Name')
plt.ylabel('2022 Data')
plt.title('Combined Plot for Top 10 Asian Countries in 2022')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Show the legend
plt.legend()

# Show the plot
plt.show()

```

```python
plt.figure(figsize=(10, 6))
sns.pointplot(x='Country Name', y='2022', data=top_10_asian_countries_2022, color='purple')
plt.xlabel('Country Name')
plt.ylabel('2022 Data')
plt.title('Point Plot for Top 10 Asian Countries in 2022')
plt.xticks(rotation=45, ha='right')
plt.show()

```

```python
# Replace these column names with your actual column names

# Selecting relevant columns
selected_columns = ['Country Name', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']
selected_data = top_10_asian_countries_2022[selected_columns]

# Computing the correlation matrix
corr_matrix = selected_data.corr()

# Plotting the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".1f", linewidths=.5)
plt.title('Correlation Heatmap for Top 10 Asian Countries (2000-2022)')
plt.show()


```
